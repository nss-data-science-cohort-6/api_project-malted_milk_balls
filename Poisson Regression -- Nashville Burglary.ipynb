```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
tract_merge = pd.read_csv('../api_project-malted_milk_balls/tract_merge.csv')
```

```python
tract_merge.head()
```

```python
income_incidents = tract_merge[['incident_number', 'median_income']]
```

```python
income_incidents['incident_number'] = income_incidents['incident_number'].astype('string')
```

```python
incident_counts_by_median_income = income_incidents.groupby('median_income')['incident_number'].count()
```

```python
incident_counts_by_median_income.plot();
```

```python
tract_merge['rate'] = tract_merge['total'] / (tract_merge['Population'] / 1000)
```

```python
tract_merge_pivot = (tract_merge.pivot_table(index = 'TRACTCE', values = ['median_income', 'Population', 'total', 'rate']))
```

```python
tract_merge_pivot
```

```python
incident_by_med_income = smf.glm('rate ~ median_income',
                 data = tract_merge,
                 family = sm.families.Poisson()
                 ).fit()

incident_by_med_income.summary()
```

```python
fit_df = pd.DataFrame({
    'median_income': np.linspace(start = tract_merge['median_income'].min(),
                       stop = tract_merge['median_income'].max(),
                       num = 150)
})
fit_df['fitted'] = incident_by_med_income.predict(fit_df)

fit_df.plot(x = 'median_income', y = 'fitted', figsize = (10,6))

fontsize = 14
plt.xlabel('median_income', fontsize = fontsize)
plt.ylabel('Estimated Mean Incident Rate', fontsize = fontsize);
```
