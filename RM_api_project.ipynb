```python
import requests
import matplotlib.pyplot as plt
import json
import pandas as pd
```

```python
with open('keys.json') as fi:
    credentials = json.load(fi)
api_key = credentials['api_key']
```

```python
url = 'https://api.census.gov/data/2020/acs/acs5/subject?get=NAME,S0101_C01_001E&for=tract:*&in=state:47&in=county:037&key={0}'.format(api_key)
```

```python
response = requests.get(url)
```

```python
response
```

```python
# Function that sends data to dataframe
def json_to_df(response):
    
    return pd.DataFrame(response.json()[1:], columns=response.json()[0])
```

```python
res=json_to_df(response)
```

```python
res.columns = ['Name', 'Population', 'state', 'county', 'tract']
res
```

```python
res['Population'] = res['Population'].astype(int)
res.info()
```

```python
Total = res['Population'].sum()
print(Total)
```

```python
url2 = 'https://api.census.gov/data/2020/acs/acs5/subject?get=NAME,S1901_C01_012E&for=tract:*&in=state:47&in=county:037&key={0}'.format(api_key)
```

```python
response2 = requests.get(url2)
response2
```

```python
res2=json_to_df(response2)
res2
```

```python
res2.columns = ['Name', 'median_income', 'state', 'county', 'tract']
res2
```

```python
census_tracts = pd.merge(res, res2, how='inner')
census_tracts
```

```python
census_tracts['median_income'] = census_tracts['median_income'].astype(int)
census_tracts.info()
```

```python
census_tracts = census_tracts[census_tracts['median_income'] > 0]
census_tracts
```

```python
census_tracts.to_csv('census_tracts.csv')
```
